```python
# 13 . Total contributation hour for each and every agents weekly basis
df4 = pd.read_csv(r'C:\Users\DELL\Desktop\Desktop\Data Science\Notes\PANDAS\Data set for pandas class 2\AgentLogingReport.csv')
df4['C_Date'] = pd.to_datetime(df4['Date'])
df4['Week'] = df4['C_Date'].dt.week
df4['Agent_Name_Week']=df4['Agent']+df4['Week'].astype(str)
```

```python
df4['Duration_Time'] = pd.to_datetime(df4['Duration'])
```

```python
df4['Time_hrs'] = df4['Duration_Time'].dt.hour + ((df4['Duration_Time'].dt.minute)/60) + ((df4['Duration_Time'].dt.hour)/3600)
df4.groupby('Agent_Name_Week')['Time_hrs'].sum()
```

```python
# 14. total percentage of active hour for a month
# perccentage is calculated w.r.t the number of working days.
# daily working hrs considered as 8 hrs
df5 = df4.groupby(['Agent','C_Date']).sum()
df5['count'] = 1
df6 = df5.groupby(['Agent']).sum()
df6['%_Active_Hrs'] = df6['Time_hrs']/(df6['count']*8)
df6
```
